<h3>Let's attempt to produce the sin wave for a complete wave cycle.</h3>

```python
a = np.arange(0, 361)  #The end-point was intentionally set a point higher since the larger value is exclusive.
b = np.sin(a)
plt.plot(a, b)     #Plotting a as x and b as y.
```

The above code produces an output. But, we certainly know that this representation is false. Eg. sin(90)=1, but we have something else. Mathematically, we need to convert the angles into radians. This is true for the numpy trig applications too. Let's re-try this.

```python
a = np.arange(0, 361)  #The end-point was intentionally set a point higher since the larger value is exclusive.
b = np.sin( np.deg2rad(a) )   # Conver from degree to radians before applying the sin function
plt.plot(a, b)     #Plotting a as x and b as y.
```

This looks perfect. Can we try same for the cosine wave, and then add other plot features.
--


```python
a = np.arange(0, 361)  #The end-point was intentionally set a point higher since the larger value is exclusive.
b = np.sin( np.deg2rad(a) )   # Conver from degree to radians before applying the sin function
c = np.cos( np.deg2rad(a) )   # Conver from degree to radians before applying the cos function
plt.plot(a, b, label='sin')     #Plotting a as x and b as y.
plt.plot(a, c, label='cos')     #label assigns a label to the plot. To appear as part of the legend, it must ne called.

plt.legend()                    #Produce a legend for the plot
plt.xlabel('angles')            #Provide a label for the x-axis
plt.ylabel('trig values')            #Provide a label for the y-axis
plt.title('The sin and cos curves', fontsize=14)                       #Provide a title for the plot
plt.xlim(0,360)                 #Sets the lower and upper limits for the xticks.
plt.ylim(-1,1)
```

```python
# Creating annotated heatmaps for NOAA ENSO data
# link/URL for data:  https://origin.cpc.ncep.noaa.gov/products/analysis_monitoring/ensostuff/ONI_v5.php

Data = np.loadtxt('data/ENSO.csv', delimiter=',', skiprows=1)         #Load Data
seasons = np.loadtxt('data/ENSO.csv', delimiter=',', max_rows=1, dtype='str')[1:] #Load first row as headers separately

years = np.int_(Data[:,0])     #Extract the years from the first column. np.int_  converts the whole array into an integer.
ENSO = Data[:,1:]              #Extract the data column and pass to the variable ENSO


fig, ax = plt.subplots(figsize=(12, 12))
im = ax.imshow(ENSO)

# show all ticks
ax.set_xticks(np.arange(len(seasons)))
ax.set_yticks(np.arange(len(years)))

# label ticks with the respective list entries
ax.set_xticklabels(seasons)
ax.set_yticklabels(years)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")


# Loop over data dimensions and create text annotations.
for i in range(len(years)):
    for j in range(len(seasons)):
        text = ax.text(j, i, ENSO[i, j], ha="center", va="center", color="w", fontsize=12)

ax.set_title("Seasonal ENSO Oceanic Nio Index")     #Set Plot Title
fig.colorbar(
    im, ax=ax, shrink=0.7, extend='both', label="ENSO INO"
)  # shrink function set the legend height; extend shows the extension arrow of either the major or minor value in the colorbar

#plt.savefig('ENSO-ONI.jpg')
```

# cartopy

`cartopy` is a Python package designed for geospatial data processing in order to produce maps and other geospatial data analyses.

Cartopy makes use of the powerful PROJ.4, NumPy and Shapely libraries and includes a programmatic interface built on top of Matplotlib for the creation of publication quality maps. To put geographical regions on the map `cartopy` includes at set of "features" (such as country boundaries, ocean, etc), but it is also possible to load your own regions using so-called "shape-files". These can be loaded with the `shapereader` function inside of `cartopy.io`.

Coordinate reference systems (CRS) are available in cartopy as `cartopy.crs` (and are inherited from proj4).


https://scitools.org.uk/cartopy/docs/latest/ 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cf
import cartopy.io.shapereader as shapereader

from pathlib import Path
```

```python
data_path = Path("data")
```

# Plotting values at stations on a map

We will here create a plot where we can see for 22 synoptic weather stations across Ghana what the annual daily maximum rainfall (ADM) was for 2020. In addition we will be adding the adding the borders of administrative regions in Ghana.

```python
drr = np.genfromtxt(data_path / "2020_ADM_rr.csv", delimiter=",", skip_header=1, usecols=[1,2,3])
stn = np.genfromtxt(data_path / "2020_ADM_rr.csv", delimiter=",", skip_header=1, usecols=0, dtype='str')

lon = drr[:, 0]
lat = drr[:, 1]
rr = drr[:, 2]
```

```python
lon
```

next we use `shapereader` to read in the "shape-file" for Ghana's administrative regions

```python
shp_path = data_path / "gh_shp/gh_wgs16dregions.shp"
adm1_shapes = list(shapereader.Reader(str(shp_path)).geometries())
```

```python
fig = plt.figure(figsize=(10,10))
# create axes with specific proojection
ax = plt.axes(projection=ccrs.PlateCarree())

#Alternative 
#fig, ax = plt.subplots( figsize=(10,10),subplot_kw={'projection': ccrs.PlateCarree()})


# set the domain lat lon extent
ax.set_extent([-4, 2, 4, 12], crs=ccrs.PlateCarree())

# add the administrative borders from the shapefile
ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), edgecolor='black', facecolor='none', alpha=0.8)

# and add country borders, ocean, land and lake borders from cartopy's set of features
ax.add_feature(cf.BORDERS)
ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(cf.LAKES, edgecolor='black')


ax.set(title='ADM Rainfall distribution - 2020')
splot = ax.scatter(lon,lat,200,c=rr, cmap=cm.jet, alpha=1)
fig.colorbar(splot,label="Rainfall (mm)", orientation="vertical", extend='both')
```
